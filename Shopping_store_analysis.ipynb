# Answer the Following questions

### what is the average age per state ?

```python
df.groupby('state')['age'].mean().round(2)
```

### Which Day has the highest number of orders ?

```python
df.groupby('day')['total_price'].sum().sort_values(ascending=False)
```

### Which Month has the highest revenue ?

```python
df.groupby('month')['total_price'].sum().sort_values(ascending=False)
```

### What is the average age per state per each gender ?

```python
df.groupby(['state', 'gender'])['age'].mean()
```

### What is the most sold product type ?

```python
df.groupby('product_type')['total_price'].sum().head(1)
```

### What is the most expensive product ?

```python
df.groupby('product_name')['total_price'].max().head(1)
```

### What is the most sold product ?

```python
df.groupby('product_name')['total_price'].sum().head(1)
```

### What is the most sold size in Shirt category ?

```python
shirt = df[df.product_type == 'Shirt']
shirt.groupby('size')['total_price'].sum().sort_values(ascending=False).head(1)
```

### What is the most sold color in Jacket category ?

```python
jacket = df[df.product_type == 'Jacket']
jacket.groupby('colour')['total_price'].sum().sort_values(ascending=False).head(1)
```

### What is the Total available quantity of Trousers in April ?

```python
trousers_april = df[(df.month == 'April') & (df.product_type == "Trousers")]
total_quantity = trousers_april['Products_Quantity'].sum()
print(f"Total Available Quantity of Trousers in April is {total_quantity}")
```

### What are the Top 10 Cities that return the highest revenue in March ?

```python
month = df[df.month == 'March']
month.groupby('city')['Sales_Quantity'].sum().sort_values(ascending=False).head(10)
```
