# Import Necessary Libraries

```python
import numpy as np
import pandas as pd
```

# Data Loading

```python
customers = pd.read_csv('customers.csv')
customers
```

```python
orders = pd.read_csv('orders.csv')
orders
```

```python
products = pd.read_csv('products.csv')
products
```

```python
sales = pd.read_csv('sales.csv')
sales
```

```python
# merge --> Join dataframes based on common column
# join --> Join dataframes based on dataframe index
```

```python
df = customers.merge(orders, how= 'inner', on= 'customer_id')
df.head()
```

```python
df = df.merge(sales, how= 'inner', on= 'order_id')
df.head()
```

```python
df = df.merge(products, how= 'inner', left_on= 'product_id', right_on= 'product_ID')
df.head()
```

```python
# In order to show all dataframe columns
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.rename(columns= {'quantity_x' : 'Sales_Quantity', 'quantity_y' : 'Products_Quantity'}, inplace= True)
```

# Data Cleaning

### For Categroical Columns

```python
cat_cols = df.select_dtypes(include= 'object').drop(['customer_name', 'home_address', 'country', 'description'], axis= 1).columns
cat_cols
```

```python
for col in cat_cols:
    print(col)
    print(df[col].nunique())
    print(df[col].unique())
    print('-' * 100)
```

```python
df.order_date = pd.to_datetime(df.order_date)
df.order_date
```

```python
df.delivery_date = pd.to_datetime(df.delivery_date)
df.delivery_date
```

```python
# Drop Unnecessary Columns
df.drop(['customer_id', 'customer_name', 'home_address','zip_code', 'country', 'sales_id', 'product_id', 'product_ID', 'description'], axis= 1, inplace= True)
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(ignore_index= True, inplace= True)

# Abother way to reset index
# df.reset_index(inplace= True, drop= True)
```

```python
df
```

```python
# Extract Year
df['year'] = df.order_date.dt.year
df['year']
```

```python
# Extract Month
# df.order_date.dt.month
df['month'] = df.order_date.dt.month_name()
df['month']
```

```python
# Extract Day
# df.order_date.dt.day
df['day'] = df.order_date.dt.day_name()
df['day']
```
